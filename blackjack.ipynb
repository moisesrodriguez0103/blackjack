```python
import random
import time

KARTEN = ["2", "3", "4", "5", "6", "7", "8", "9", "10", "Bube", "Dame", "König", "Ass"] 

def kart_ziehhen():
    return random.choice(KARTEN)

def hand_anzeigen(name, hand, verdeckt=False):
    if verdeckt:
        print(f"{name}'s Hand: [X, {hand[0]}]")

    else:
        print(f"{name}'s Hand: {hand} (Wert: {hand_wert(hand)})")

def hand_wert(hand):
    wert = 0
    as_count = 0
    for k in hand:
        if k in ["Bube", "Dame", "König"]:
            wert += 10
        elif k == "Ass":
            wert += 1
            as_count += 1
        else:
            wert += int(k)
    # Ass kann 1 oder 11 sein, also prüfen, ob es besser ist, 11 zu nehmen
    if as_count > 0 and wert + 10 <= 21:
        wert += 10
    return wert

def blackjack():
    spieler = [kart_ziehhen(), kart_ziehhen()]
    dealer = [kart_ziehhen(), kart_ziehhen()] 

    print("\n ==== Blackjack ===")
    hand_anzeigen("Dealer", dealer, verdeckt=True)
    hand_anzeigen("Spieler", spieler)

    while True:
        wert_spieler = hand_wert(spieler)
        if wert_spieler == 21:
            print("Spieler hat Blackjack!")
            break
        if wert_spieler > 21:
            print("Spieler hat verloren (Über 21)")
            return
        wahl = input("hit oder stand? (h/s) ").strip().lower()
        if wahl == 'h':
            spieler.append(kart_ziehhen())
            hand_anzeigen("Spieler", spieler)
        elif wahl == 's':
            break
        else:
            print("Ungültige Eingabe. Bitte gib 'h' für hit oder 's' für stand ein.")

    # Dealer zieht Karten, bis er 17 oder mehr hat
    print("\nDealer deckt auf...")
    time.sleep(2)  # Kurze Pause, um Spannung zu erzeugen
    hand_anzeigen("Dealer", dealer)
    while hand_wert(dealer) < 17:
        print("Dealer zieht eine Karte...")
        dealer.append(kart_ziehhen())
        time.sleep(2)
        hand_anzeigen("Dealer", dealer)

    wert_spieler = hand_wert(spieler)
    wert_dealer = hand_wert(dealer)

    if wert_spieler > 21:
        print("Spieler hat verloren (Über 21)")
    elif wert_dealer > 21:
        print("Spieler gewinnt (Dealer über 21)")
    elif wert_spieler > wert_dealer:
        print("Spieler gewinnt!")
    elif wert_spieler < wert_dealer:
        print("Dealer gewinnt!")
    else:
        print("Unentschieden!")

def main():
    while True:
        blackjack()
        nochmal = input("\nMöchtest du nochmal spielen? (j/n) ").strip().lower()
        if nochmal == 'n':
            print("Danke fürs Spielen! Auf Wiedersehen!")
            break 
        elif nochmal != 'j':
            print("Ungültige Eingabe. Bitte gib 'j' für ja oder 'n' für nein ein.")
        
main()        
```
